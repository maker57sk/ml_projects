```python
import pandas as pd
import datetime
import plotly.express as px
import plotly.graph_objects as go
from pandas.core.tools.datetimes import to_datetime
```

```python
def data_loading(data):
  df  = pd.read_csv(data)
  return df
```

```python
def convert_date_columns(df):
  if df["Connected On"].dtype == 'datetime64[ns]':
    return df
  elif df["Connected On"].dtype == 'O':
    df['Connected On'] = pd.to_datetime(df['Connected On'])
    return df
```

```python
def pipeline(data):
  df = data_loading(data)
  df = convert_date_columns(df)
  return df
```

```python
def connections_plot(df):
  x="Connected On"
  y="First Name" 
  labels={'First Name': 'Number'}
  title='Connections in my LinkedIn Network'

  connections_line = px.line(df.groupby(by=x).count().reset_index(), 
                           x=x, 
                           y=y, 
                           labels=labels,
                           title=title)
  return connections_line.show()
```

```python
def positions_plot(df, start, end):
  x='Position'
  y='First Name'
  labels={'First Name': 'Number'}
  title= 'Positions in my LinkedIn Network'
  ascending=False
  
  return px.bar(df.groupby(by=x ).count().sort_values(by=y,
                                                      ascending=ascending)[start:end].reset_index(),
                                                      x=x,
                                                      y=y,
                                                      labels=labels,
                                                      title=title)
```

```python
df = pipeline('Connections.csv')
connections_plot(df)
positions_plot(df, 0, 50)
```
