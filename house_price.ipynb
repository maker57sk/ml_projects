```python
# Importing dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics 
```

```python
# Importing boston house price dataset from sklearn
boston_dataset = sklearn.datasets.load_boston()
```

```python
boston_dataset
```

```python
# Converting dataset to pandas dataframe

df = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
```

```python
df.head()
```

```python
# Adding price column to dataframe form target variable

df['price'] = boston_dataset.target
```

```python
df.head()
```

```python
# number of rows and columns
df.shape
```

```python
# number of missing values
df.isnull().sum()
```

```python
# Statistical Measures
df.describe()
```

```python
correlation = df.corr()
```

```python
correlation
```

```python
# heatmap for understanding correlation

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
# Seprating features and labels

X = df.drop('price', axis=1)
Y = df['price']
print(X)
print(Y)
```

```python
# Spliting into train and test data

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# defining the model
model = XGBRegressor()
```

```python
# training the model using train dataset
model.fit(X_train, Y_train)
```

```python
# predicting using train dataset
train_prediction = model.predict(X_train)
```

```python
# measuring accuracy with R squared error on train data
score_1_train = metrics.r2_score(Y_train, train_prediction)
# measuring accuracy with MAE 
score_2_train = metrics.mean_absolute_error(Y_train, train_prediction)
print('R2: {}\nMAE: {}'.format(score_1_train,score_2_train))
```

```python
# ploting actual Price vs Predicted Price on training data
plt.scatter(Y_train, train_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price on training data')
plt.show()
```

```python
# prediction on test data
test_prediction = model.predict(X_test)
# measuring accuracy with R squared error on test data
score_1_test = metrics.r2_score(Y_test, test_prediction)
# measuring accuracy with MAE on test data
score_2_test = metrics.mean_absolute_error(Y_test, test_prediction)
print('R2: {}\nMAE: {}'.format(score_1_test,score_2_test))
```

```python

plt.scatter(Y_test, test_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price on testing data')
plt.show()
```
