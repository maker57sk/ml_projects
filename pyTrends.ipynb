```python
from pytrends.request import TrendReq
```

```python
lang = 'en-US'
timez = 360
topics = ["crypto", "data science", "Blockchain"]
```

```python
pytrends = TrendReq(hl=lang, tz=timez)
pytrends.build_payload(kw_list=topics)
df = pytrends.interest_over_time()
df.plot(figsize=(10,7))
```
